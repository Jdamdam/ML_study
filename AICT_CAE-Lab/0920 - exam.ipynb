```python
import tensorflow as tf
import numpy as np
tf.set_random_seed(777)  

x_exam = np.loadtxt('exam2.csv', usecols=(0,1), delimiter=',', dtype=np.float)
y_exam = np.loadtxt('exam2.csv', usecols=(3), delimiter=',', dtype=np.float)
x_test1= np.loadtxt('test2.csv', usecols=(0,1), delimiter=',', dtype=np.float)
y_test1= np.loadtxt('test2.csv', usecols=(3), delimiter=',', dtype=np.float)

x_data = x_exam[:, 0:-1]
y_data = y_exam[:, 0:-1]
X_test = x_test1[:, [-1]]
X_test = y_test1[:, [-1]]

# placeholders for a tensor that will be always fed.
X = tf.placeholder(tf.float32, shape=[None, 2])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([2, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

# Hypothesis using sigmoid: tf.div(1., 1. + tf.exp(tf.matmul(X, W)))
func = tf.matmul(X, W) + b #weight와 data, bias를 더한 값 => 성적기준
hypothesis = tf.sigmoid(tf.matmul(X, W) + b)

# cost/loss function
cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) * tf.log(1 - hypothesis))

train = tf.train.GradientDescentOptimizer(learning_rate=0.25).minimize(cost)

# Accuracy computation
# True if hypothesis>0.5 else False
predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))


with tf.Session() as sess:
    # Initialize TensorFlow variables
    sess.run(tf.global_variables_initializer())

    for step in range(10001):
        cost_val, train_val, func_val = sess.run([cost, train, func], feed_dict={X: x_data, Y: y_data})
        if step % 200 == 0:
            print("\nStep: ",step," Cost: ",cost_val, "\nWeight \n",sess.run(W), "\nFunction_value \n", func_val*100)

    # Accuracy report
    f, h, c, a = sess.run([func, hypothesis, predicted, accuracy],
                       feed_dict={X: x_data, Y: y_data})
    print("\nFunction\n", f*100, "Hypothesis \n", h, "\nCorrect (Y) \n ", c, "\nAccuracy \n ", a)
    
    
    t_f, t_h, t_c, t_a = sess.run([func, hypothesis, predicted, accuracy],
                       feed_dict={X: X_test, Y: Y_test})
    print("\ntest_Function \n", t_f*100, "\ntest_Hypothesis \n", t_h, "\ntest_Correct (Y) \n", t_c, "\ntest_Accuracy \n", t_a)
    
```
