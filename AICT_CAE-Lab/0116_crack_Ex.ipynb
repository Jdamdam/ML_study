```python
import tensorflow as tf
import numpy as np
tf.set_random_seed(777)  

x_data = np.loadtxt('crack_or_cicle_training2.csv', usecols=(0,1), delimiter=',', dtype=np.float)
y_data = np.loadtxt('crack_or_cicle_training2.csv', usecols=(2), delimiter=',', dtype=np.float)
y_data = y_data.reshape(-1,1)

x_test= np.loadtxt('crack_or_cicle_test.csv', usecols=(0,1), delimiter=',', dtype=np.float)
y_test= np.loadtxt('crack_or_cicle_test.csv', usecols=(2), delimiter=',', dtype=np.float)
y_test = y_test.reshape(-1,1)

#print('x_data\n', x_data, '\ny_data\n', y_data, '\nx_test\n', x_test, '\ny_test\n', y_test)

#x_data = x_data[:, 0:-1]
#y_data = y_data[:, 0:-1]
#X_test = x_test[:, [-1]]
#X_test = y_test[:, [-1]]

#print('x_data\n', x_data, '\ny_data\n', y_data, '\nx_test\n', x_test, '\ny_test\n', y_test)
```

```python
X = tf.placeholder(tf.float32, shape=[None, 2])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W1 = tf.Variable(tf.random_normal([2, 10]), name='weight1')
b1 = tf.Variable(tf.random_normal([10]), name='bias1')
layer1 = tf.nn.relu(tf.matmul(X, W1) + b1)
#layer1 = tf.nn.relu(tf.matmul(X, W1) + b1)

W2 = tf.Variable(tf.random_normal([10, 10]), name='weight1')
b2 = tf.Variable(tf.random_normal([10]), name='bias1')
layer2 = tf.nn.relu(tf.matmul(layer1, W2) + b2)

W3 = tf.Variable(tf.random_normal([10, 1]), name='weight1')
b3 = tf.Variable(tf.random_normal([1]), name='bias1')
hypothesis = tf.sigmoid(tf.matmul(layer2, W3) + b3)
hypothesis = tf.clip_by_value(hypothesis,1e-5,1-(1e-5))
# cost/loss function
cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) * tf.log(1 - hypothesis))

optimizer = tf.train.AdamOptimizer(learning_rate=0.1).minimize(cost)

# Accuracy computation
# True if hypothesis>0.5 else False
predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))

#sess.run(tf.global_variables_initializer())

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        cost_val, train_val, acc, _ = sess.run([cost, optimizer, accuracy, hypothesis], feed_dict={X: x_data, Y: y_data})
        if step % 100 == 0:
           # print("\nStep: ",step," Cost: ",cost_val, "\nWeight \n",sess.run(W))
            print("Training Step: {:5}\tLoss: {:.5f}\tAcc: {:.2%}".format(step, cost_val, acc))
    
    pred = sess.run(predicted, feed_dict={X: x_data})

   # for p, y in zip(pred, y_data.flatten()):
   #     print("[{}] Prediction: {} True Y: {}".format(p == y, p, y))
    
    print("----------------------------------------------------------------------------------")    
    print("------------------------------------TEST--SET-------------------------------------")   
    print("----------------------------------------------------------------------------------")    
    
    pred = sess.run(predicted, feed_dict={X: x_test})
    
    # y_data: (N,1) = flatten => (N, ) matches pred.shape
    for p, y in zip(pred, y_test.flatten()):
        print("[{}] Prediction: {} True Y: {}".format(p == y, p, y))
    
#     # Accuracy report
#     _, c, a = sess.run([hypothesis, predicted, accuracy],
#                        feed_dict={X: x_data, Y: y_data})
#     print("\nCorrect (Y) \n ", c, "\nAccuracy \n ", a)
    
    
#     _, t_c, t_a = sess.run([hypothesis, predicted, accuracy],
#                        feed_dict={X: x_test, Y: y_test})
#     print("\ntest_Correct (Y) \n", t_c, "\ntest_Accuracy \n", t_a)
```
